```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=np.array([10,25])
y=np.array([0,350])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'+')
plt.show()
# Marker Type
#'o' 	Circle 	
#'*' 	Star 	
#'.' 	Point 	
#',' 	Pixel 	
#'x' 	X 	
#'X' 	X (filled) 	
#'+' 	Plus 	
#'P' 	Plus (filled) 	
#'s' 	Square 	
#'D' 	Diamond 	
#'d' 	Diamond (thin) 	
#'p' 	Pentagon 	
#'H' 	Hexagon 	
#'h' 	Hexagon 	
#'v' 	Triangle Down 	
#'^' 	Triangle Up 	
#'<' 	Triangle Left 	
#'>' 	Triangle Right 	
#'1' 	Tri Down 	
#'2' 	Tri Up 	
#'3' 	Tri Left 	
#'4' 	Tri Right 	
#'|' 	Vline 	
#'_' 	Hline

#Line Type
#'-' 	Solid line 	
#':' 	Dotted line 	
#'--' 	Dashed line 	
#'-.' 	Dashed/dotted line

#Colors
#'r' 	Red 	
#'g' 	Green 	
#'b' 	Blue 	
#'c' 	Cyan 	
#'m' 	Magenta 	
#'y' 	Yellow 	
#'k' 	Black 	
#'w' 	White
```

```python
x=np.array([1,20,5])
y=np.array([12,22,4])
plt.plot(x,y)
plt.show()
```

```python
sizes=np.array([100,50,20])
plt.scatter(x,y,color="green")
plt.scatter(x,y,color="yellow",s=sizes,alpha=0.3)
plt.show()
# plt.scatter(x,y,co)
```

```python
plt.bar(x,y,color="green",width=2)
```

```python
plt.barh(x,y,color="green",height=1)
plt.show()
```

```python
plt.hist(x)
```

```python
plt.pie(x)
```

```python
track_stud=np.array([20,30,40,20,10])
track_name=["Data","Mobile","Web","IoT","CyberSecurity"]
ex=[0.2,0,0,0,0]
col=["Red","Pink",'m','g','b']
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=col)
plt.legend(title="Tracks")
plt.show()
```
